```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 15

y = np.linspace(-2,2,n)
t = np.linspace(-2,2,n)

Y, T = np.meshgrid(y,t)
f = Y - 2

angulo = np.arctan(f)

dy = np.cos(angulo)
dt = np.sin(angulo)

plt.quiver(T,Y,dy,dt)
```

```python
f = 2 - Y

angulo = np.arctan(f)

dy = np.cos(angulo)
dt = np.sin(angulo)

plt.quiver(T,Y,dy,dt)
```

```python
f = - Y - 2

angulo = np.arctan(f)

dy = np.cos(angulo)
dt = np.sin(angulo)

plt.quiver(T,Y,dy,dt)
```
